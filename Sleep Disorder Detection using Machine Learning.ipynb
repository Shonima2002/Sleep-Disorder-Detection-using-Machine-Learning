## To transform non numerical labels to numerical labels

## Split the column 'Blood Pressure' into Upper_BP	Lower_BP

```python
df[['Upper_BP', 'Lower_BP']] = df['Blood Pressure'].str.split('/', expand=True)
```

## Heatmap

## Split the columns into input and target variables

```python
x = df.drop(columns = 'Sleep Disorder')
y = df['Sleep Disorder']
```

```python
x.head()
```

## Carry out feature scaling using standard scalar

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled=scaler.fit_transform(x)
```

```python
x = pd.DataFrame(x_scaled, columns=x.columns)
```

```python
x.head()
```

## Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3, random_state= 100)
```

## Machine Learning Models

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
print("Accuracy score of our model is : ", metrics.accuracy_score(y_test,y_pred))

print("Confusion matrix is : ", '\n' , metrics.confusion_matrix(y_test,y_pred))

print("Classification report is : ", '\n', metrics.classification_report(y_test,y_pred))

print("Cross Validation Score is: ", cross_val_score(LogisticRegression(), x, y, cv = 2).mean())
```

### Visualize confusion matrix

```python
sns.heatmap(metrics.confusion_matrix(y_test,y_pred), annot = True, fmt = 'd')
plt.show()
```
